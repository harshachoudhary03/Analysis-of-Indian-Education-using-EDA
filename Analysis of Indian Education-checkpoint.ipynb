## Analysis of Indian Education System using EDA  

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('dropout-ratio-2012-2015.csv')
df1
```

```python
df1.columns
```

```python
df1.isnull()
```

```python
df1.isna().sum()
```

```python
west_bengal = df1[df1['State_UT']=="West Bengal"]
west_bengal.head()
```

```python
df1['Primary_Boys'].values
```

```python
df1 = df1.replace('NR',0)
```

```python
df1['Primary_Boys'] = df1['Primary_Boys'].apply(lambda x: float(x))
```

```python
df1['Primary_Girls'] = df1['Primary_Girls'].values.astype(np.float32)
```

```python
df1['Upper Primary_Boys'].values
```

```python
df1 = df1.replace('Uppe_r_Primary',0)
```

```python
df1['Upper Primary_Boys'] = df1['Upper Primary_Boys'].values.astype(np.float32)
df1['Upper Primary_Girls'] = df1['Upper Primary_Girls'].values.astype(np.float32)
```

```python
df1['Secondary _Boys'] = df1['Secondary _Boys'].values.astype(np.float32)
df1['Secondary _Girls'] = df1['Secondary _Girls'].values.astype(np.float32)

df1['HrSecondary_Boys'] = df1['HrSecondary_Boys'].values.astype(np.float32)
df1['HrSecondary_Girls'] = df1['HrSecondary_Girls'].values.astype(np.float32)
```

```python
df1.describe()
```

```python
df1.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize= (35,15))
sns.countplot(df1['State_UT'])
```

```python
plt.bar(west_bengal['Primary_Girls'],west_bengal['year'])
plt.xlabel("Primary Girls")
plt.ylabel("Year")
plt.title("Year based primary girls rstio in west bengal")
plt.show()
```

```python
plt.bar(west_bengal['Primary_Boys'],west_bengal['year'])
plt.xlabel("Primary Boys")
plt.ylabel("Year")
plt.title("Year based primary boys rstio in west bengal")
plt.show()
```

```python
plt.bar(df1['Secondary _Girls'],df1['year'])
plt.xlabel("Secondary Girls")
plt.ylabel("Year")
plt.title("Year based Secondary Girls rstio in west bengal")
plt.show()
```

```python
plt.bar(df1['Secondary _Boys'],df1['year'])
plt.xlabel("Secondary Boys")
plt.ylabel("Year")
plt.title("Year based Secondary boys ratio in west bengal")
plt.show()
```
